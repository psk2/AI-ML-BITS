```python
df = pd.read_csv('jiomart_groceries.csv')
df.info()
```

```python
#Checking for null or missing values in my data
df.isnull().sum()
```

```python
df["Discount"] = df["Discount"].str.replace("%","")

df
```

```python
df["Original_Price"] = df.loc[:,"Original_Price"].str.replace("₹","").str.replace(",","").str.strip()

```

```python
#Changing data types for column Discounted_Price, Discount, Original_Price
convert_dict = {'Discounted_Price' : float, 
                'Discount' : float,
                'Original_Price' : float,
                'No_of_ratings' : int,
               }

df = df.astype(convert_dict) 
print(df.info())
```

```python
# Find a duplicate rows
duplicateDFRow = df[df.duplicated(['Product_name'])]
print(duplicateDFRow)
```

```python
# For the ease sake and to work with other's data, creating one more csv for integration.
df.to_csv('jio_mart_data_cleaned.csv', index=False, header=True)
```
